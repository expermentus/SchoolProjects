```python
import math
import itertools

def basic_counting():
    print("\n##############################################################\nA set of events where all events are equally likely to happend\n##############################################################\n")
    print("if the question is asked what the propability that an event, A, which is a subset of S, is going to happen.")
    print("Remember that for a finite sample space S with equally likely outcomes, the probability of an event A is given by: P(A)=|A|/|S|= M/N")
    option = int(input("\n1: input size of the set and subset\n2: input values of the set\n Option: "))
    
    match option:
        
        case 1:
            
            S = float(input("\nInput size of S (the superset): "))
            A = float(input("\nInput size of A (the subset): "))
            Probability = A / S
            
        case 2:
                  
            S = set()
            A = set()
            i = 1
            
            print(f"\ninput elements of S (the superset)\ntype 'stop' to finish the inputs\n")
            while True:
                s = input(f"S[{i}]: ")
                if s == 'stop':
                    i = 1
                    break
                i += 1
                S.add(s)
            
            print(f"\ninput elements of A (the subset)\ntype 'stop' to finish the inputs\n")
            while True:
                a = input(f"A[{i}]: ")
                if a == 'stop':
                    break
                i += 1
                A.add(a)
    
            if A.issubset(S) == True:
                Probability = len(A)/len(S)
                
            else:
                count = 0
                for number in S:
                    if number in A:
                        count += 1
                        
                Probability = count/len(S)
                
        case _:
            print('invalid input')
            return 0
                
    print(f"\n In this case, the probability of event A happening would be: {Probability} or {Probability * 100}% \n\n")
    return Probability

def ordered_sampling_with_replacement():
    print("\n###################################################################################################\nA set of events where all events are equally likely to happend, ordered sampling with replacement\n###################################################################################################\n")
    print("We have a set with n elements (e.g.: A={1,2,3,⋯.n}), and we want to draw k samples from the set such that ordering matters and repetition is allowed. For example, if A={1,2,3}and k=2, there are 9 different possibilities:")
    print("(1,1), (1,2), (1,3), (2,1), (2,2), (2,3), (3,1), (3,2), (3,3).")
    print("In general, we can argue that there are k positions in the chosen list: ( Position 1 , Position 2 , ..., Position k) . There are n options for each position. Thus, when ordering matters and repetition is allowed, the total number of ways to choose k objects from a set with n elements is n*n*...*n=n^k. \nNote that this is a special case of the multiplication principle where there are k 'experiments' and each experiment has n possible outcomes.")
    
    A = set()
    i = 1
    
    print(f"\ninput elements of A\ntype 'stop' to finish the inputs\n")
    while True:
        a = input(f"A[{i}]: ")
        if a == 'stop':
            break
        i += 1
        A.add(a)
    
    print(f"\ninput k (the amount of samples to fetch from the set) \n")
    k = int(input("k: "))
    
    n = int(len(A))
    
    Possibilities = n ** k
    total = 0
    print(f"\nIn this case, there are {Possibilities} ways to choose k samples from n")
    while True: 
        results_option = input("\nWould you like to review all possibilities?\n1: Yes\n2: No\nOption: ")   
        match results_option:
            case '1':
                print("The possibilities are as stated:")  
                
                # for all possibilities, ordered, replacements
                combinations = itertools.product(sorted(list(A)), repeat = k)
                
                for combination in combinations:
                    print(combination)
                break  
                
            case '2':
                break
            
            case _:
                print("\n#############\ninvalid input\n#############")
                continue
        
    print(f"\nTo calculate the possibility, you would need to consider the amount of events and then divide it by the possibilities ( {total} / {Possibilities} )")

def ordered_sampling_without_replacement():
    print("\n######################################################################################################\nA set of events where all events are equally likely to happend, ordered sampling without replacement\n######################################################################################################\n")
    print("We have a set with n elements (e.g.: A={1,2,3,⋯.n}), and we want to draw k samples from the set such that ordering matters and repetition is not allowed. For example, if A={1,2,3}and k=2, there are 6 different possibilities:")
    print("(1,2), (1,3), (2,1), (2,3), (3,1), (3,3).")
    print("In general, we can argue that there are k positions in the chosen list: ( Position 1 , Position 2 , ..., Position k) . \nThere are n options for the first position, (n-1) options for the second position (since one element has already been allocated to the first position and cannot be chosen here), (n-2) options for the third position, ... (n-k+1) options for the k th position.\nThis formular can be derived to: n!/(n-k)! \nThus, when ordering matters and repetition is not allowed, the total number of ways to choose k objects from a set with n elements is")
    
    A = set()
    i = 1
    
    print(f"\ninput elements of A\ntype 'stop' to finish the inputs\n")
    while True:
        a = input(f"A[{i}]: ")
        if a == 'stop':
            break
        i += 1
        A.add(a)
    
    print(f"\ninput k (the amount of samples to fetch from the set) \n")
    k = int(input("k: "))
    
    n = int(len(A))
    
    # p(n choose k) = n! / (n-k)!:
    Possibilities = math.factorial(n) / math.factorial(n - k)
    
    Possibilities = int(Possibilities)
    
    print(f"\nIn this case, there are {Possibilities} ways to choose k samples from n")
    while True: 
        results_option = input("\nWould you like to review all possibilities?\n1: Yes\n2: No\nOption: ")   
        match results_option:
            case '1':
                print("The possibilities are as stated:")  
                
                # for all possibilities, ordered, replacements
                combinations = itertools.permutations(sorted(list(A)), k)
                
                for combination in combinations:
                    print(combination)
                break  
            
            case '2':
                pass
                break
            
            case _:
                print("\n#############\ninvalid input\n#############")
                continue
        
    print(f"\nTo calculate the possibility, you would need to consider the amount of events and then divide it by the possibilities ( |Events| / {Possibilities} )")
    
def unordered_sampling_without_replacement():
    print("\n######################################################################################################\nA set of events where all events are equally likely to happend, unordered sampling without replacement\n######################################################################################################\n")
    print("We have a set with n elements (e.g.: A={1,2,3,⋯.n}), and we want to draw k samples from the set such that ordering does not matters and repetition is not allowed.\nThus, we basically want to choose a k -element subset of A , which we also call a k -combination of the set A . For example if A={1,2,3} and k=2 , there are 3 different possibilities::")
    print("(1,2), (1,3), (2,3).")
    print("We show that number of k-element subsets of A by P(n chooses k)\n A typical scenario here is that we have a group of n people, and we would like to choose k of them to serve on a committee. Note that the difference between the two is ordering. In fact, for any k -element subset of A={1,2,3,....n} , we can order the elements in k! ways, thus we can write")
        
    A = set()
    i = 1
    
    print(f"\ninput elements of A\ntype 'stop' to finish the inputs\n")
    while True:
        a = input(f"A[{i}]: ")
        if a == 'stop':
            break
        i += 1
        A.add(a)
    
    print(f"\ninput k (the amount of samples to fetch from the set) \n")
    k = int(input("k: "))
    
    n = int(len(A))
    
    # n choose k = n! / k!(n-k)!:
    Possibilities = math.factorial(n) / math.factorial(k) * (math.factorial(n - k))
    
    Possibilities = int(Possibilities)
    
    print(f"\nIn this case, there are {Possibilities} ways to choose k samples from n")
    while True: 
        results_option = input("\nWould you like to review all possibilities?\n1: Yes\n2: No\nOption: ")   
        match results_option:
            case '1':
                print("The possibilities are as stated:")  
                
                # for all possibilities, unordered, replacements
                combinations = itertools.combinations(sorted(list(A)), k)
                
                for combination in combinations:
                    print(combination)
                break  
            
            case '2':
                pass
                break
            
            case _:
                print("\n#############\ninvalid input\n#############")
                continue
        
    print(f"\nTo calculate the possibility, you would need to consider the amount of events and then divide it by the possibilities ( |Events| / {Possibilities} )")
    
def unordered_sampling_with_replacement():
    print("\n######################################################################################################\nA set of events where all events are equally likely to happend, unordered sampling with replacement\n######################################################################################################\n")
    print("uppose that we want to sample from the set A={a1,a2,...,an} k times such that repetition is allowed and ordering does not matter. For example, if A={1,2,3} and k=2 , then there are 6 different ways of doing this:")
    print("(1,1), (1,2), (1,3), (2,2), (2,3), (3,3)")
    print("Calculating all possibilities: One way to think about this is to note that any of the pairs in the above list can be represented by the number of 1 's, 2 's and 3 's it contains. That is, if x1 is the number of ones, x2 is the number of twos, and x3 is the number of threes, we can equivalently represent each pair by a vector (x1,x2,x3)\n i.e., 1,1 →(x1,x2,x3)=(2,0,0)\n1,2 →(x1,x2,x3)=(1,1,0)\n1,3 →(x1,x2,x3)=(1,0,1)\n2,2 →(x1,x2,x3)=(0,2,0)\n2,3 →(x1,x2,x3)=(0,1,1)\n3,3 →(x1,x2,x3)=(0,0,2).")
    print("The formular for calculating this is n+k-1 choose k, or (n+k-1)! / k!(n-1)!")
        
    A = set()
    i = 1
    
    print(f"\ninput elements of A\ntype 'stop' to finish the inputs\n")
    while True:
        a = input(f"A[{i}]: ")
        if a == 'stop':
            break
        i += 1
        A.add(a)
    
    print(f"\ninput k (the amount of samples to fetch from the set) \n")
    k = int(input("k: "))
    
    n = int(len(A))
    
    # n+k-1 choose k = (n+k-1)! / k!(n-k)!:
    Possibilities = math.factorial(n + k - 1) / (math.factorial(k) * (math.factorial(n - 1)))
    
    Possibilities = int(Possibilities)
    
    print(f"\nIn this case, there are {Possibilities} ways to choose k samples from n")
    while True: 
        results_option = input("\nWould you like to review all possibilities?\n1: Yes\n2: No\nOption: ")   
        match results_option:
            case '1':
                print("The possibilities are as stated:")  
                
                # for all possibilities, ordered, replacements
                combinations = itertools.combinations_with_replacement(sorted(list(A)), k)
                
                for combination in combinations:
                    print(combination)
                break  
            
            case '2':
                pass
                break
            
            case _:
                print("\n#############\ninvalid input\n#############")
                continue
        
    print(f"\nTo calculate the possibility, you would need to consider the amount of events and then divide it by the possibilities ( |Events| / {Possibilities} )")
```

-	Cumulative distribution function is calculated like this:

```python
import numpy as np
import matplotlib.pyplot as plt

# Input the discrete values
values = [1, 2, 3, 4, 5, 5, 6, 7, 8, 8, 8, 9, 9, 10]

# Compute the CDF
counts, bin_edges = np.histogram(values, bins=np.arange(0, 11))
cdf = np.cumsum(counts) / len(values)

# Plot the CDF
plt.step(bin_edges[1:], cdf)
plt.xlabel('Value')
plt.ylabel('CDF')
plt.show()
```

```python
import matplotlib.pyplot as plt
import collections

def generate_pmf():
    print("\nProbability mass function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value")
    print("Formular for PMF is: Px(x) = P(X=x) (small x and big X are different variables) = Σ(x): PX(x)=1")
    print("All event can either be equally likely, or not")
    print("Are all events equally likely?\n1: Equally likely\n2: Not equally likely")
    option = input("Option: ")
    
    match option:
        
        case '1':
            
            print("Define elements of the set, or just the length?\n 1: Define lenght of the set (Events called a, b ... z)\n 2: Define elements of the set")
            option = input("Option: ")
            
            match option:
                
                case '1':
                    length = int(input("Lenght of the set (must be an int): "))
                    # Define the sample set
                    A = set()

                    for i in range(length):
                        A.add(chr(ord('`')+i+1))    
                    
                    A  = sorted(A)
                    # Count the number of occurrences of each object in the set
                    freq_dict = dict(collections.Counter(A))

                    # Calculate the probability of each object in the set
                    n = len(A)
                    prob_dict = {x: freq_dict[x] / n for x in freq_dict}
                    
                    # Define the range of possible outcomes
                    outcomes = range(1, n)

                    # Create lists to store the objects and their probabilities
                    outcomes = list(prob_dict.keys())
                    probs = list(prob_dict.values())
                    
                    print("\nProbability of each event is:", round(probs[0],2), "or", round(probs[0]*100,2),"%")

                    # Create a bar plot of the PMF
                    plt.bar(outcomes, probs,width = 0.025)
                    plt.xlabel('Events')
                    plt.ylabel('Probability')
                    plt.title('Probability Mass Function')
                    plt.show()
                
                case '2':
                    # Define the set of possible outcomes
                    A = set()
                    i = 1
                    
                    print(f"\ninput elements of A\ntype 'stop' to finish the inputs\n")
                    while True:
                        a = input(f"A[{i}]: ")
                        if a == 'stop':
                            break
                        i += 1
                        A.add(a)
                        
                    # Count the number of occurrences of each object in the set
                    freq_dict = dict(collections.Counter(A))

                    # Calculate the probability of each object in the set
                    n = len(A)
                    prob_dict = {x: freq_dict[x] / n for x in freq_dict}
                    
                    # Define the range of possible outcomes
                    outcomes = range(1, n)
                    
                    # Create lists to store the objects and their probabilities
                    outcomes = list(prob_dict.keys())
                    probs = list(prob_dict.values())
                    
                    print("\nProbability of each event is:", round(probs[0],2), "or", round(probs[0]*100,2),"%")
                    
                    # Create a bar plot of the PMF
                    plt.bar(sorted(list(A)), 1/len(A), width = 0.025)
                    plt.xlabel('Outcome')
                    plt.ylabel('Probability')
                    plt.title('Probability Mass Function')
                    plt.show()
                    plt.cla()
            
        case '2':

            # Define the set of possible outcomes
            A = set()
            i = 1
            prob_dict = {}
            total = 0

            print(f"\ninput elements of A\ntype 'stop' to finish the inputs\n")
            while True:
                a = input(f"A[{i}]: ")
                if a == 'stop':
                    break
                probability = float(input(f"probability of A[{i}]: "))
                prob_dict[a] = probability
                total += probability
                i += 1
                A.add(a)

            if total > 1:
                print("\n#############\ninvalid input\n#############:\nThe probability of all events can max be 1\n")    

            else:
                # Create lists to store the outcomes and their probabilities
                outcomes = list(prob_dict.keys())
                outcomes.sort(key=lambda x: float(x) if isinstance(x, (int, float)) else x)
                probs = [prob_dict[outcome] for outcome in outcomes]

                # Create a bar plot of the PMF
                plt.bar(outcomes, probs)
                plt.xlabel('Outcome')
                plt.ylabel('Probability')
                plt.title('Probability Mass Function')
                plt.show()
                plt.cla()

generate_pmf()
```

Probability density function:

we define the normal_pdf function that takes three inputs: the value(s) of x, the mean value mu, and the standard deviation sigma. We then use the np.linspace function to generate an array of 1000 equally spaced values between -5 and 5 to represent the range of x values over which we want to evaluate the PDF. 
We set mu = 0 and sigma = 1 to define a standard normal distribution. 

We use the normal_pdf function and the x array to generate an array of PDF values. 

Finally, we use Matplotlib to plot the input range against the PDF values. The resulting plot shows a bell-shaped curve that represents the PDF of the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))

x = np.linspace(-5, 5, 1000)
mu = 0
sigma = 1

pdf = normal_pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Normal Distribution PDF')
plt.show()
```

Types of variables:

Pmf:
P{X=i}=(■(n@i)) p^i (1-p)^(n-i)

We do something until we have a success
Pmf:
P{M=k}=(1-p)^(k-1) p

The hypergeometric random variable:
we have objects of two types: type 1 N objects (acceptable components) and type 2 M objects (defective components). A sample of n objects is randomly chosen without replacement.
Pmf:
P{X=i}=((N/i)*(M/n-i)))/((N+M)/n)
E[X]=nN/(N+M)
Var(X)=np(1-p)[1-(n-1)/(N+M-1)]
